```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def create_heatmap_matrix(csv_name):
  csv_name = 'exp_comparison.csv'
  df = pd.read_csv(csv_name, usecols=[0,1,2], names=['Experiment 1 Player', 'Experiment 2 Player', "Experiment 1's Wins"])
  df = pd.pivot_table(df, index='Experiment 2 Player', columns='Experiment 1 Player', values="Experiment 1's Wins")

  sns.set(rc={"figure.figsize":(9, 12)})
  # colors = ["#fc8d59", "#ffffbf", "#91cf60"]
  # color_scheme = sns.color_palette(colors)
  color_scheme = sns.color_palette("vlag_r", 3)
  # color_scheme = sns.dark_palette("seagreen", as_cmap=True)
  g = sns.heatmap(df, annot=True, cmap=color_scheme, linewidths=1, linecolor="white", square=True, cbar=False)
  g.invert_yaxis()
```

```python
create_heatmap_matrix('exp_comparison.csv')
```
